```python
from scipy import stats
import numpy as np
```

```python
z_score = np.abs(stats.zscore(df['tran_amount']))
threshold = 3
outliers = z_score > threshold
print(df[outliers])
```

```python
z_score = np.abs(stats.zscore(df['response']))
threshold = 3
outliers = z_score > threshold
print(df[outliers])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x = df['tran_amount'])
plt.show()
```

```python
sns.boxplot(x = df['response'])
plt.show()
```

```python
df['month'] = df['trans_date'].dt.month
```

```python
df
```

```python
M_sales = df.groupby('month')['tran_amount'].sum()
M_sales = M_sales.sort_values(ascending = False).reset_index()
M_sales
```

```python
C_count = df['customer_id'].value_counts().reset_index()
C_count.columns = ['customer_id','count']
C_count
```

```python
top5 = C_count.sort_values(by = 'count',ascending = False).head(5)
top5
```

```python
sns.barplot(x = 'customer_id', y = 'count', data = top5)
```

```python
C_sales = df.groupby('customer_id')['tran_amount'].sum().reset_index()
C_sales

top5_sales = C_sales.sort_values(by = 'tran_amount',ascending = False).head(5)
top5_sales
```

```python
sns.barplot(x = 'customer_id', y = 'tran_amount', data = top5_sales)
```
